### Stripping white space

Later we will learn how to process text files with Python. To do so, we can strip the white spaces so we can process the string. Those white spaces include
- ' ' or '\s'- spaces
- '\n' - new lines
- '\t' - tabs

```python
astring = "     Python is a versatile, readable language  .\t... \n"
# .strip() takes out all white spaces from the ends
print("original")
print(astring)
print("*")
print("stripped")
print(astring.strip())
print("*")
```

```python
astring = "     Python is a versatile, readable language  .\t... \n"
print(astring.lstrip())
print("*")
```

```python
astring = "     Python is a versatile, readable language  .\t... \n"
print(astring.rstrip())
print("*")
```

```python
par = "Marie Curie was a remarkable scientist. \n She discovered radium and polonium, and her research on radioactivity changed science forever."

def count_words_in_paragraph(par):
    par = par.lower()   
    par = par.replace("\n", " ") 
    for char in ",.?":
      par = par.replace(char, " ")            
    words = par.split()               
    word_dict = {}                    

    for word in words:                
        if word in word_dict:
            word_dict[word] += 1
        else:
            word_dict[word] = 1

    return word_dict

print(count_words_in_paragraph(par))

   
```

```python
def main():
    ''' get user input and test if it is a palindrome'''
    user_input = input("Enter a string to check if it's a palindrome: ")
    lowercase_input = user_input.lower()
    reversed_words = ""
    for i in range(len(lowercase_input)-1, -1, -1):
        reversed_words += lowercase_input[i]
    if reversed_words == lowercase_input:
        print(f"'{lowercase_input}' is a palindrome!")
    else:
        print(f"'{lowercase_input}' is not a palindrome!")
main()

        


```

```python
def main():
    '''Creates a dictionary from area codes and city names'''

    # List of area codes for cities in the US
    area_codes = [212, 213, 312, 305, 415]

    # List of corresponding cities
    cities = ["New York City, NY", "Los Angeles, CA", "Chicago, IL", 
              "Miami, FL", "San Francisco, CA"]

    # Create a dictionary from the two lists
    area_dict = dict(zip(area_codes, cities))

    # Print the dictionary
    print(area_dict)


main()
```

```python
class Dog:
    def __init__(self, name, color, breed, age):
        '''Initialize a Dog with name, color, breed, and age'''
        self.name = name
        self.color = color
        self.breed = breed
        self.age = age

    def display_info(self):
        '''Displays the information of the Dog'''
        print(f"Dog's name: {self.name}")
        print(f"Color: {self.color}")
        print(f"Breed: {self.breed}")
        print(f"Age: {self.age}")
        print("-" * 20)

    def change_age(self, new_age):
        '''Change the age of the Dog'''
        self.age = new_age
        print(f"{self.name}'s age has been updated to {self.age}.")


def main():
    '''Create and use two Dog instances'''
    # Ask user for dog details
    name1 = input("Enter first dog's name: ")
    color1 = input("Enter its color: ")
    breed1 = input("Enter its breed: ")
    age1 = int(input("Enter its age: "))

    name2 = input("Enter second dog's name: ")
    color2 = input("Enter its color: ")
    breed2 = input("Enter its breed: ")
    age2 = int(input("Enter its age: "))

    # Create Dog objects
    dog1 = Dog(name1, color1, breed1, age1)
    dog2 = Dog(name2, color2, breed2, age2)

    # Display both dogs
    dog1.display_info()
    dog2.display_info()

    # Change age of one dog
    new_age = int(input(f"Enter new age for {dog1.name}: "))
    dog1.change_age(new_age)

    # Display again
    dog1.display_info()


main()
```
